```python
from itertools import permutations 

n = int(input())
nums = list(map(int,input().split()))
opers = list(map(int,input().split()))
operator  = []
for idx,item in enumerate(opers):
    for j in range(item):
        if idx == 0:
            operator.append("+")
        elif idx == 1:
            operator.append("-")
        elif idx == 2:
            operator.append("*")
        elif idx == 3:
            operator.append("//")
            
permOP = list(permutations(operator, n-1))
sumList = []

# print("permOP : ", permOP)
# numSum = nums[0]
for n in range(len(permOP)):
    numSum = nums[0]
    for a in range(len(permOP[n])):
        num = nums[a+1]
        op = permOP[n][a]
        
        print("numSum : ",numSum)
        print("op : ",op)
        print("num : ", num)
        
        if op == "+":
            numSum += num
        elif op == "-":
            numSum -= num
        elif op == "*":
            numSum *= num
        elif op == "//":
            numSum //= num
        else:
            pass
        if n == 66:
            print("herherehrherehr")
            print(numSum)
#         print("-----")
    sumList.append(numSum)
    
#     print(sumList)
print(max(sumList))
print(min(sumList))
```
