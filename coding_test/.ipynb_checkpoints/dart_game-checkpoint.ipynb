```python
def solution(dartResult):
    num = int()
    numList = []
    operList = []
    start = False
    answer = 0
    priceCount = 0
    
    dartResult = dartResult.replace("10", "K")
    
    #numList에 담기.
    for index, item in enumerate(dartResult):
        if item == "K":
            num = 10
            numList.append(num)
        elif item.isdigit():
            num = int(item)
            numList.append(num)
        else :
            operList.append(item)
            
    newIndex = 0
    
    for index, oper in enumerate(operList) :
        if oper == "S":
            continue
            
        elif oper == "D":
            newIndex = index - priceCount
            numList[newIndex] **= 2
            
        elif oper == "T":
            newIndex = index - priceCount
            numList[newIndex] **= 3
            
        elif oper == "*":
            priceCount += 1
            newIndex = index - priceCount
            if newIndex > 0 :
                numList[newIndex] *= 2
                numList[newIndex-1] *= 2
            elif newIndex == 0 :
                numList[newIndex] *= 2
                
        elif oper == "#":
            priceCount += 1
            newIndex = index - priceCount
            numList[newIndex] *= (-1)

    answer = sum(numList)
    
    return answer
```
