```python
n, m = list(map(int,input().split()))
tree = list(map(int,input().split()))
st = 1
end = max(tree)
sumTree = 0
while st < end :
    sumTree = 0
    height = (st + end)//2
    print("I am height",height)
    for t in tree :
        if t > height:
            sumTree += t - height
    print("I am sumTree", sumTree)
    if sumTree < m :
        end = height-1
    else :
        st = height +1
        
print(height)
```
