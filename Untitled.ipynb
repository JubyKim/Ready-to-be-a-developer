```python
def isPalindrome(head: list):
    leftPoint = 0
    rightPoint = len(head)-1
    
    while leftPoint < rightPoint:
        if head[leftPoint] == head[rightPoint]:
            print(head[leftPoint])
            print(head[rightPoint])
            leftPoint += 1
            rightPoint -= 1
            continue
        else :
            return False
    return True
```

```python
N, M = map(int, input().split())  # M = depth
result = []
visitied = [False]*(N+1)

def dfs(depth, N, M):
    if depth == M+1:  # 깊이까지 탐색 완료
        for i in range(0,len(result)):
            print(result[i],end=" ")
        print()
        return
    for num in range(1,N+1):
        if not visitied[num]:
            visitied[num] = True
            result.append(num)
            dfs(depth+1, N, M)
            visitied[num] = False
            result.pop()

dfs(1,N,M)
```

```python
n,m,s = map(int,input().split())
graph = {str(i) :[] for i in range(1,n+1)}

# queue = []
# stack = []
# dfsVisit = []
# bfsVisit = []

for i in range(m):
    one, two = input().split()
    graph[one].append(two)
    
for i,j in graph.items():
    graph[i] = sorted(j)

def dfs(graph, start_node):
    visit = list()
    stack = list()

    stack.append(start_node)

    while stack:
        node = stack.pop()
        if node not in visit:
            visit.append(node)
            stack.extend(graph[node])
    return visit

def bfs(graph, start_node):
    visit = list()
    queue = list()

    queue.append(start_node)

    while queue:
        node = queue.pop(0)
        if node not in visit:
            visit.append(node)
            queue.extend(graph[node])

    return visit

print(" ".join(dfs(graph,str(s))))
print(" ".join(bfs(graph,str(s))))
```

```python
#등차수열
import itertools
n = int(input())
numList = list(map(int,input().split()))
count = 0
answerList = []

for idx,item in enumerate(numList):
    for jdx,jtem in enumerate(numList):
        if idx != jdx :
            center = int((item + jtem) / 2)
            if center in numList and abs(item - center)== abs(jtem - center):
                answerList.append((item,center,jtem))
                
# for a in range(len(answerList)):
#     answerList[a] = tuple(sorted(answerList[a]))
# print(answerList)
answerList = set(answerList)

permList = list(itertools.permutations(numList,3))
for i in answerList :
    if i in permList:
#         print(i)
        count += 1
print(count-2)
```
